# **Line Plot**

Line plot is a type of data visualization that is widely used and is the basic type of plot in data visualization. This type of plot displays information in the form of a series of data points connected by straight line segments. Line plots are used for various fields. Data points can be represented by symbols or straight lines without icons. Line plots are used to visualize data trends in time intervals.

Line plots can be used on datasets that have continuous values to see the movement of data over time. To create a line plot, we can use the **.plot()** function.

```python
data.plot()
```

It can be seen that only the salary data has a clear trend, while the PhD and service data are not shown clearly in the plot. This is because the numerical units in salary reach hundreds of thousands, while PhD and services are in very small units. For this reason, visualization must use more comparable data such as PhD and service data.

```python
data[["phd","service"]].plot()
```

There is a correlation between PhD and service over the years, except from the ages of 70 to 80, where service declines, meaning some people leave service at age 70 and over.

# **Bar Plot**

Bar Plot is one of the most frequently used in data visualization. Bar plots are used to represent data horizontally. The vertical lengths of the bars represent the values. These bar plots are used to represent a single data series and data points related to groups of data in a series.

To make a bar plot, we can use the **bar()** function from Matplotlib. In this example, we will create a bar plot of the salary data. 

```python
data['salary'].plot.bar()
```

The image above shows the output bar plot of all salary data. We can also visualize bar plots by setting the desired amount of data by adding the **head()** or **tail()** functions. For example, we will make a bar plot of the top 10 data from the salary data.

```python
data['salary']. head(10).plot.bar()
```

If the example output above is in the form of a vertical bar plot, then we can also visualize a bar plot in a horizontal form. Making horizontal bar plots is not much different from vertical bar plots. The difference is only in the method used. To create a horizontal bar plot we use the **barh()** function. For example, we will create a horizontal bar plot of the first 10 data points of the salary data.

```python
data['salary'].head(10).plot.barh()
```

In addition to visualizing bar plots with one data variable, we can also visualize bar plots for 2 variable data at once. In this case, what needs to be considered is that the data used must have comparable units. For example, we will create a bar plot visualization of the phd and service data for the top 10 data.

```python
 data[[ 'phd', 'service' ]].head(10).plot.bar()
```

# **Pie Chart**

Pie charts are also known as pie charts or pie charts. The pie chart is divided into sectors, each circle representing a proportion of the total. Pie charts are used to show size as a data comparison. To visualize the distribution of percentages, a pie chart (pie chart) works best.

To create a pie chart, we can use the **pie()** function. For example, we will make a pie chart of rank data for comparison of each type of rank.



```python
rank = ['Prof', 'AssocProf', 'AsstProf']
jumlah_rank = ['46', '13', '19']

fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.pie(jumlah_rank, labels = rank, autopct='%1.0f%%')
ax.set_title('Diagram Rank')
```

It can be seen that the highest rank percentage is Prof rank at 59%, followed by AsstProf rank at 24%, and AssocProf rank at 17%.

# **Box Plot**
Box plots are an important visualization that helps us understand the distribution of quartiles of numerical data. Box plots or box-whisker plots are concise representations to help understand the various quartiles, slope, dispersion, and outliers in data. Box plots are used to compare variables using multiple statistical values. Comparable variables must be from the same data unit. The boxplot summarizes the distribution of each attribute, drawing a line for the median (middle value) and boxes around the 25th and 75th percentiles (middle 50% of the data). Whiskers provides an overview of the distribution of data and points outside the whiskers indicate outlier candidate values (values that are 1.5 times larger than the middle 50% spread size of the data).

To create a boxplot, we can use the **boxplot()** function. For example, we will create a boxplot of salary data.

```python
 data[['salary']].boxplot()
```

We can also create a boxplot visualization of two variable data. For example, we will create a boxplot visualization of the phd and service data to compare the median and quartiles.

```python
data[["phd","service"]].boxplot()
```

It can be seen that the median and quantiles of phd are higher than the median and quantiles of services. In addition, the range of PhD is wider than the range of services.

# **Histogram Plot**

From this form, it can quickly be seen whether an attribute is Gaussian, skewed or even has an exponential distribution. The histogram can also help to spot possible outliers.

To create a histogram, we can use the **hist()** function.

# **Scatter Plot**

Scatter plots are another type of visualization that are usually used to identify correlations or patterns among attributes. Scatter plot shows the relationship between two experimental factors (eg phd and service). Trendlines are used to determine positive, negative, or no correlation.

To make a scatter plot we can use the **scatter()** function from matplotlib. The scatter() function is to plot one point on each observation. This function takes two arrays of equal length, one for the x-axis values, and one for the y-axis values.

For example, we will create a scatter plot visualization with phd data as values on the x-axis and service data as values on the y-axis.

```python
x = data['phd']
y = data['service']

plt.scatter(x, y)
```

# **Density Plot**

Density Plots are another way to get a quick idea of the distribution of each attribute. The density plot looks like an abstract histogram with a smooth curve drawn through the top of each bar.

Density plots can be created using the **density()** function. The **density()** function does not generate a density graph directly. The **density()** function only calculates the kernel density of the data. Next, for plotting, we use the **plot()** function.

The following is the syntax and the resulting output in the image below.



```python
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
plt.show()
```

# **Strip Plot**

Strip Plot is a categorical data visualization method for numeric data. Strip plots are also known as 1-dimensional scatter plots or dot plots. This Visualization method is good for seeing the distribution of data values
based on data categories (data groups) with a relatively small amount of data so that this method can also be an alternative visualization other than box plots with a small amount of data.

To create a strip plot, we use the **stripplot()** function from the Seaborn library. With stripplot() we can create scatterplot graphs based on categories. For example, we will create a strip plot visualization with sex data as the x-axis value (category) and salary data as the y-axis value.

```python
sns.stripplot(x='sex', y='salary', data=data)
```

The image above visualizes the salary variable based on gender (sex) category.

# **Violin Plot**

Violin plot is a combination of two methods, namely boxplot and Kernel Density Estimation (KDE). KDE itself is a useful way to analyze non-parametric data by calculating the probability density function of a random variable. The purpose of the violin plot is to make it easier for users to analyze the continuous distribution of data for each category. In accordance with KDE, the more convex the violin plot data graph is visualized, the greater the probability of data density. Conversely, the flatter the violin plot data graph visualized, the smaller the chance of data density. The quartile and whisker values are included in the violin plot.

To create a violin plot, we use the violinplot() function from the Seaborn library. For example, we will create a violin plot visualization with rank data as the x-axis value and salary data as the y-axis value, to visualize salary data based on rank type.

```python
sns.violinplot(x='rank', y='salary', data=data)
```

Based on the picture above, it is known that the salary data for rank type AsstProf has the highest probability density, which can be seen from the violin plot graph for Asstprof which is more convex than the graph for ranks AssocProf and Prof. While the smallest opportunity density is in the Prof type salary data, which is shown by the violin plot graph which is flatter than the other two graphs.

# *Correlation Matrix Plot**

Correlation gives an indication of how related changes are between two variables. If two variables change in the same direction they are positively correlated. If they change in opposite directions together (one up, one down), then they are negatively correlated. The correlation between each pair of attributes is called the correlation matrix. The correlation matrix can be plotted to get an idea of which variables have a high correlation with each other. This is important to know, because some machine learning algorithms such as linear and logistic regression can have poor performance if there are highly correlated input variables in your data.

The following is the syntax and output of the correlation matrix plot which is presented in the image below.

```python
names = ["rank", "discipline","phd","service","sex", "salary"]
correlations = data.corr()

#Correlation Matrix Plot
fig, ax = plt.subplots()
fig.set_size_inches((8,10))
sns.heatmap(data.corr(), square=True, annot=True)
```
